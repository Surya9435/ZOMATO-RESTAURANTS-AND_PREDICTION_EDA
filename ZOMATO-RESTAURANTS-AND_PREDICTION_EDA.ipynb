# Cleaning of Rate Column

```python
df['rate'].unique()
```

# Removing "NEW,"-" and "/5" from rate Column

```python
#As we can see there is 'NEW','nan','-' present in the column and also the denomenators, so we have to remove this.

def correct_rate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value= str(value).split('/')
        value= value[0]
        return float(value)

df['rate']=df['rate'].apply(correct_rate)
df['rate'].head()
```

# Now I will check that whether NULL VALUES are present in the 'rate' Column

```python
df.rate.isnull().sum()
```

# Filling Null Values in 'rate' Column with MEAN

```python
df['rate']=df['rate'].fillna(df['rate'].mean())
df.rate.isnull().sum()
```

# Now again I will check different columns by .info

```python
df.info()
```

```python
# Checking the Null Values for 'location' column

df.location.isnull().sum()
```

```python
# The Presence of Null Values is very low & so I will Drop the Null Values

df=df.dropna()
df.head()
```

```python
df.location.unique()
```

```python
df["listed_in(city)"].unique()
```

# Listed_in(city) & Location both are same type of columns, so lets keep one

```python
df=df.drop(["listed_in(city)"],axis=1)
```

# Checking the unique values present in "approx_cost(for two people)" column

```python
df["approx_cost(for two people)"].unique()
```

# Removing (,)COMMA ) from "approx_cost(for two people)" Column

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].apply(handlecomma)
df['approx_cost(for two people)'].unique()
```

```python
df.head()
```

# Cleaning rest_type Column

```python
rest_types = df["rest_type"].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

# Now i will make a cluster of the rest_types_less than 1000 as Others

```python
def handle_rest_type(value):
    if (value in rest_types_lessthan1000):
        return "others"
    else:
        return value
    
df["rest_type"] = df["rest_type"].apply(handle_rest_type)
df["rest_type"].value_counts()
```

```python
df.head()
```

# Cleaning location Column

```python
df["location"].value_counts()
```

# Now i will make a cluster of the location less than 300 as Others

```python
location = df["location"].value_counts(ascending = False)

location_lessthan300 = location[location<300]

def handle_location(value):
    if (value in location_lessthan300):
        return "others"
    else:
        return value

df["location"] = df["location"].apply(handle_location)
df["location"].value_counts()
```

```python
df.head()
```

# Cleaning cuisines Column

```python
cuisines = df["cuisines"].value_counts(ascending = False)

cuisines_lessthan100 = cuisines[cuisines<100]

def handle_cuisines(value):
    if (value in cuisines_lessthan100):
        return "others"
    else:
        return value
    
df["cuisines"] = df["cuisines"].apply(handle_cuisines)
df["cuisines"].value_counts()
```

```python
df["listed_in(type)"].value_counts()
```
